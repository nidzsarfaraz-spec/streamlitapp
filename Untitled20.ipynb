```python
import pandas as pd
```

```python
df=pd.read_csv("sports_data.csv")
```

```python
df
```

```python
df["country"]=df["Player"].str.split(pat="(").str[1]
df["country"]=df["country"].str.split(pat="(").str[0]
df["Player"]=df["Player"].str.split(pat="(").str[0]
df["start_year"]=df["Span"].str.split(pat="=").str[0].astype(int)
df["end_year"]=df["Span"].str.split(pat="=").str[1].astype(int)
df.rename(columns={"Span":"Duration", "Mat":"Matches","BF":"Balls_faced","HS":"High_score", "SR":"Strike_rate"},inplace=True)
df["Duration"]=df["end_year"]-df["start_year"]
df["High_score"]=df["High_score"].str.split(pat="*").str[0].astype(int)
df["Balls_faced"]=df["Balls_face"].str.split(pat="+").str[0].astype(int)
df["Balls_faced"]=df["Balls_faced"].replace("-",0)
df["Balls_faced"]=df["Balls_faced"].astype(float)
df["Strike_rate"]=df["Strike_rate"].replace("-",0)
df["Strike_rate"]=df["Strike_rate"].astype(float)
df["0"]=df["0"].replace("-",0)
df["0"]=df["0"].astype(int)
```
