```python
log=[]
def admin():
    while True:
        no=int(input("Enter a number of user you want to create:"))
        for i in range(no):
            username=input(f"Enter a user {i+1} name:").lower()
            password=input(f"Enter a {i+1} Password:")
            log.append([username,password])
        check=input("What you want to check: \n1:Count of users \n2:want to search any user? \n3:Wnat to remove any user? \n4:want to add any user?")
        if check=="1":
            print(f"Total number of users are: {len(log)}")
            break
        elif check=="2":
            lst=[]
            for i in range(len(log)-1):
                lst.append(log[i][0])
            checkk2=input("Enter a name of user you want to search:").lower()
            count=lst.count(check2)
            if count>0:
                print("user exist!")
                break
            else:
                print("user does not exist!")
                break
        elif check=="3":
            lst=[]
            for i in range(len(log)):
                lst.append(log[i][0])
            check2=input("Enter a name of user you want to remove?").lower()
            var=lst.index(check2)
            print(log.pop(var))
            print(f"updated User details are :{log}")

        break
```
