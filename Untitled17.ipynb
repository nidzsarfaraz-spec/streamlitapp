```python
df.rename(columns={"Player":"Players", "Span":"Duration", "Mat":"Matches", "Inns":"Innings", "NO":"Not_out","HS":"High_score", "Ave":"Average", "BF":"Balls_faced", "SR":"Strike"},inplace=True)
```

```python
df.info()
```

```python
df["country"]=df['Players'].str.split(pat="(").str[1]
```

```python
df
```

```python
df["country"]=df["country"].str.split(pat=")").str[0]
```

```python
df
```

```python
df["Players"]=df["Players"].str.split(pat="(").str[0]
```

```python
df
```

```python
df["Start_year"]=df["Duration"].str.split(pat="-").str[0]
```

```python
df
```

```python
df["ending_year"]=df["Duration"].str.split(pat="-").str[1]
```

```python
df
```

```python
df["Start_year"]=df["Start_year"].astype(int)
df["ending_year"]=df["ending_year"].astype(int)
```

```python
df.info()
```

```python
df["Duration"]=df["ending_year"]-df["Start_year"]
```

```python
df.info()
```

```python
df
```

```python
df["Balls_faced"]=df["Balls_faced"].str.split(pat="+").str[0]
```

```python
df
```

```python
df.info()
```

```python
 df["High_score"]=df["High_score"].str.split(pat="*").str[0]
```

```python
df["Strike"]=df["Strike"].replace("-",0)
```

```python
df["Balls_faced"]=df["Balls_faced"].replace("-",0)
```

```python
df["Balls_faced"]=df["Balls_faced"].astype(int)
```

```python
df["High_score"]=df["High_score"].astype(int)
```

```python
df["Strike"]=df["Strike"].astype(float)
```

```python
df.info()
```

```python
df["0"]=df["0"].replace("-",0)
```

```python
df["0"]=df["0"].astype(int)
```

```python
df.info()
```

```python
df["Duration"]=df["Duration"].astype(object)
```

```python
df.info()
```
