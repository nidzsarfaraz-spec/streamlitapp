```python
#School Management system

roll=[]
def admin():
    nos=int(input("Enter a number of students you want to add:"))
    for i in range(nos+1):
        rollstd=f"{i}-2026-10"
        roll.append(rollstd)
    print(f"{nos} Roll Numbers has been generated successfully!")
        
            
```
